# 01 — Tokenization

LLMs read and bill in **tokens**, not characters. Here you'll see how text maps to
token counts. Vigil's `count_tokens` uses `tiktoken` when available and falls back to
a ~4-chars-per-token heuristic otherwise — so this runs fully offline.

```python
import sys
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))

from levels.l5_memory import count_tokens
```

```python
samples = [
    'CVE-2021-44228',
    'Log4Shell is a critical remote code execution vulnerability.',
    'The quick brown fox jumps over the lazy dog.' * 5,
]
for s in samples:
    print(f'{count_tokens(s):>5} tokens  |  {len(s):>5} chars  |  {s[:48]!r}')
```

## Try it yourself
Change the text below and watch the token count move. Notice that rare strings and
punctuation cost relatively more tokens than common English words.

```python
my_text = 'Replace me with your own advisory text and re-run the cell.'
print(count_tokens(my_text), 'tokens')
```

**Why it matters:** token budgets drive cost *and* context limits. This is exactly
what L5's context-window management (F5) protects against — see
`build_budgeted_memory_context` in `levels/l5_memory.py`.
