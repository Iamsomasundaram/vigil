# 02 — Prompt Engineering

The **system** message shapes the model's behaviour before it ever sees the task.
Here you'll compare how different system roles change the *intended* output — and
construct messages the same way every Vigil level does.

```python
import os, sys
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
has_key = bool(os.getenv('OPENAI_API_KEY'))

def build_messages(cve_id, system_role):
    return [
        {'role': 'system', 'content': system_role},
        {'role': 'user', 'content': f'Analyse {cve_id} and summarise its risk.'},
    ]
```

```python
roles = {
    'terse': 'You are an analyst. Answer in one sentence, no preamble.',
    'exec': 'You are briefing a non-technical executive. Avoid jargon.',
    'engineer': 'You are advising an SRE. Be precise and include remediation steps.',
}
for name, role in roles.items():
    print(f'--- {name} ---')
    print(build_messages('CVE-2021-44228', role))
    print()
```

```python
if has_key:
    from openai import OpenAI
    client = OpenAI()
    model = os.getenv('OPENAI_MODEL', 'gpt-4o-mini')
    for name, role in roles.items():
        r = client.chat.completions.create(model=model, messages=build_messages('CVE-2021-44228', role), max_tokens=120)
        print(f'### {name}\n{r.choices[0].message.content}\n')
else:
    print('No API key — compare the message lists above. The system role is the lever.')
```

**Takeaway:** same task, different system role — very different output. This is the
cheapest, highest-leverage control you have over an LLM.
