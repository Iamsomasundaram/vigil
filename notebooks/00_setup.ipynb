# 00 — Setup & Your First Call

This notebook checks your environment and (optionally) makes a first LLM call.
Everything here is **offline-friendly**: if no `OPENAI_API_KEY` is set, the live
call is skipped and you still see the message structure.

```python
import os, sys
from pathlib import Path

# Make the project importable from the notebooks/ folder.
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))

has_key = bool(os.getenv('OPENAI_API_KEY'))
print('Project root :', ROOT)
print('OPENAI_API_KEY set:', has_key)
```

## The atomic unit: a list of messages
Every level in Vigil ultimately builds a list of `{role, content}` messages.

```python
messages = [
    {'role': 'system', 'content': 'You are a senior cybersecurity analyst.'},
    {'role': 'user', 'content': 'Analyse CVE-2021-44228 and summarise its risk.'},
]
messages
```

```python
if has_key:
    from openai import OpenAI
    client = OpenAI()
    resp = client.chat.completions.create(model=os.getenv('OPENAI_MODEL', 'gpt-4o-mini'), messages=messages, max_tokens=120)
    print(resp.choices[0].message.content)
else:
    print('No API key — skipping the live call. Set OPENAI_API_KEY to try it.')
```
